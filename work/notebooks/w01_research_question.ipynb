```python
print("aggregated avg_position grouped by trend direction : ")
print(df.groupby("trend_direction")["avg_position"].mean())
print("="*60)
print("aggregated impressions_90d grouped by position_tier : ")
print(df.groupby("position_tier")["impressions_90d"].mean())
print("="*60)
print("aggregated ctr grouped by main intent :")
print(df.groupby("main_intent")["ctr"].mean())
```
